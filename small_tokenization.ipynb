```python
import pandas as pd
import numpy as np
from transformers import AutoTokenizer
```

```python
# Load a tokenizer (e.g., GPT-2, BERT, etc.)
tokenizer = AutoTokenizer.from_pretrained("gpt2")  # Try "bert-base-uncased" for WordPiece

text = "ChatGPT's tokenization splits words."
tokens = tokenizer.tokenize(text)

print(tokens)
```
